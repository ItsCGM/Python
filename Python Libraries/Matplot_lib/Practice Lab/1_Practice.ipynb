```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading data
movies_df = pd.read_csv('C:/Users/chitr/python/Python Library/Matplot_lib/netflix_movies (1).csv')
movies_df.head()
```

```python
print(movies_df.columns)
print(movies_df.info())
```

```python
# looking for the missing values
print(movies_df.isnull().sum())
```

```python
# removing unecessary column and NaN/ None Values.
movies_df.drop(['director', 'cast', 'title', 'date_added'], axis=1, inplace=True)
movies_df.dropna(inplace=True)
```

```python
movies_df.isnull().sum()
```

```python
type = movies_df['type'].value_counts()
```

```python
# Plotting a bar of No. of TV shows vs Movies

plt.figure(figsize=(2,6))
plt.bar(type.index, type.values, color=['skyblue', 'lightgreen'],width=0.5)
plt.title('No. of Movies vs TV shows of Netflix')
plt.xlabel('type')
plt.ylabel('Count')

plt.tight_layout()
plt.grid()
plt.show()
```

```python
# Ploting a Pie chart for for the rating of the movies.
rating_type = movies_df['rating'].value_counts()
print(rating_type)
plt.figure(figsize=(8,6))
plt.pie(rating_type.values,labels= rating_type.index, autopct='%1.1f%%' )
plt.title("Distribution of rating of Movies and TV_shows")
plt.axis('equal')
plt.show()
```

```python
movies_df['country'].value_counts()
```

```python
top_countires = movies_df['country'].value_counts().head(10)
```

```python
plt.figure(figsize = (10,6))
plt.bar(top_countires.index, top_countires.values)
plt.xlabel('Country Name')
plt.ylabel("Number of Movies")
plt.title("Top 10 Countries")

plt.gca()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(top_countires.index, top_countires.values, color='skyblue')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 10 Countries')
plt.xticks(rotation=45, ha='right')  # Rotate and right-align
plt.tight_layout()
plt.grid()
plt.show()
```
