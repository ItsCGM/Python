```python

month = ['Jan', 'Feb', 'Mar', 'Apr','May', 'Jun']
sales = [1000,1200,800,950,999,1150]

plt.plot(month,sales,label= "Sales Data", color="blue",linewidth = 2,linestyle='-.')


plt.title("This Year Sales") # giving a title to the Graph 
plt.xlabel("Month") # label to the X-axis
plt.ylabel("Sales (in Rupees)") # label to the Y-axis

plt.legend(loc="lower right") # determine the location of the labels
plt.show()
```

## Scatter Plot
- plt.scatter(hours_study, exam_marks, color='green', marker='^', s=30, label='Student Data')

```python
hours_study = [1,2,3,4,5,6,7,8]
exam_marks = [56,60,70,75,78,80,78,90]

plt.scatter(hours_study, exam_marks, color='green', marker='^', s=30, label='Student Data')

plt.xlabel("Hours Study")
plt.ylabel("Test Scored")
plt.title("Student Performance")
plt.ylim(40,100)
plt.legend()
plt.grid()
plt.show()
```

Changing the "font", "font color", "font size" we have a method for it called "fontdict".
- $NOTE:$ A major difference b\w 'fontweight'= "thickness of the wordL" and 'fontsize' is "size of word"


```python
x = [1,2,3,4]
y = [5,6,9,10]
plt.plot(x,y,'ro--', linewidth=2, markersize=5)
plt.title("Y vs X Graph", 
          fontdict={'fontname' : "Calibri", 'fontweight': 12,'fontsize':25, "color": 'blue'}) # giving a title to the Graph 
plt.xlabel("X-axis",
            fontdict={'fontname' : "Calibri", 'fontweight': 5,'fontsize':20, "color": 'red'}) # label to the X-axis
plt.ylabel("y-axis",
            fontdict={'fontname' : "Calibri", 'fontweight': 100,'fontsize':15, "color": 'green'}) # label to the Y-axis


plt.show()
```

## Data Visualization Tool & Their Purpose
- Bar Charts:   " plt.bar(x, y, color='color_name', width=value, label="label_name") "
    - Category Comparision
    - Data Analysis
    - $ NOTE:$ Bar charts can also be displayed horizontally. " plt.barh() "
- Pie Charts " plt.pie(values, labels="label_list", autopct='%1.1f%%') "
    - Proposonal Illustration
    - Whole Representation

- Histogram Charts " plt.hist(data, bins=number_of_bins, edgecolor="balck") "
    - Numerical Distribution
    - Data Insight

```python
x = np.array(["python",'java',"c++"])
y = np.array([85,34,50])

# c = ['b', 'r', 'g'] # we can also pass different color to each bar.
plt.bar(x,y,width = 0.2, color="y", edgecolor='yellow',linewidth=3,linestyle=':', label='No. of Student')
# for the Horizontal Representation of bar chart, "width" is replaced by "height".
plt.xlabel("Language", fontsize=15)
plt.ylabel('pularity', fontsize= 15)
plt.grid()
plt.legend()
plt.show()
```

```python
regions = ["North", 'South', 'East', 'West']
revenue = [3000, 2700, 2500, 1500]
plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['gold', 'skyblue','lightgreen', 'coral'])
plt.title("Revenue Contribution by Regions")
plt.show()
```

```python
scores = [45,56,48,85,46,25,94,39,98,25,47,86,24,85,69,59,97,94,85,79,28,58,79,55]

plt.hist(scores, bins= 5, color='skyblue',edgecolor='black', label='marks') # bins is similar as providing range to the frequecy distribution.
plt.xlabel('Score Range')
plt.ylabel('No. of Students')
plt.title("Score Distribution of Students")
plt.legend()
plt.show()
```
