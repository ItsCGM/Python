```python
import matplotlib.pyplot as plt
```

# Subs-plots 
- plt.subplot(rows, column, index)



```python
x = [1,2,3,4]
y = [10,20,15,25]

plt.subplot(1,2,1) # draws fig at index 1 in canvas
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2) # Draws fig at index 2 in canvas.
plt.bar(x,y)
plt.title("Bar Chart")

plt.tight_layout() # auto fit figures in Canvas.
plt.show()
```

Their is a proffesonal way to draw multiple figures in a same Canvas.

"fig, ax = ptl.subplots(1,2, figsize=(10,5))  # passing figsize in completly optional. "

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))

ax[0].plot(x,y)
ax[0].set_title("Line Plot")

ax[1].bar(x,y)
ax[1].set_title("Bar Plot")

fig.suptitle("Comparison of Line and Bar Chart")
plt.show()
```

# Saving figure:
- " savefig('filename.extension', dpi=value, bbox_inches='tight') "

```python
plt.plot(x, y, color='b', marker='o')
plt.title("Simple Line Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-label')

plt.savefig("line-plot.png", dpi=300,bbox_inches='tight')
plt.show()
```
